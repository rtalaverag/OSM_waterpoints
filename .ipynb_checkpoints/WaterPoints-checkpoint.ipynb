```python
import osmnx as ox
import matplotlib.pyplot as plt
import contextily as ctx
```

```python
import warnings
warnings.filterwarnings('ignore')

ox.config(log_console=True, use_cache=True)
```

### Specify the name that is used to search for the data

```python
place_name = "Barcelona, Spain"
place = place_name.split(', ')[0].upper()
```

### Retrieve water points

```python
water_tag = {'amenity':['drinking_water', 'water_point']}
water_points = ox.geometries_from_place(place_name, water_tag)
num_waterpoints = len(water_points)
water_points = water_points.to_crs(epsg=3857)
```

### Retrieve highways

```python
ways_tag = {'highway':['motorway', 'trunk', 'primary', 'secondary', 'tertiary', 'residential', 'pedestrian', 'living_streets', 'footway']}
ways = ox.geometries_from_place(place_name, ways_tag)
ways = ways.to_crs(epsg=3857)
```
